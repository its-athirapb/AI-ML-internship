```python
import numpy as np
```

```python
#creating array
a = np.array([1, 2, 3, 4, 5])
print(a)
```

```python
#2D array
b = np.array([[1, 2, 3],
              [4, 5, 6]])
print(b)
```

```python
#special arrays
print(np.zeros((3, 3)))
print(np.ones((2, 4)))
print(np.eye(3))
```

```python
#Array Attributes
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Size:", arr.size)
```

```python
#Array Operations
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition:", a + b)
print("Multiplication:", a * b)
```

```python
#Sum and Mean
arr = np.array([1, 2, 3, 4])

print("Sum:", arr.sum())
print("Mean:", arr.mean())
print("Max:", arr.max())
print("Min:", arr.min())
```

```python
#Indexing and Slicing
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])
print(arr[1:4])
```

```python
#Reshape
arr = np.arange(6)

print(arr)
print(arr.reshape(2, 3))
```

```python
#Broadcasting
a = np.array([1, 2, 3])

print(a + 5)
```

```python
# Save to a file
np.save('array.npy', arr)
```

```python
loaded_arr = np.load('array.npy')
print(loaded_arr)
```
